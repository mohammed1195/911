```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = df1.drop('e', axis=1)
```

```python
df.describe()
```

```python
df.head(5)
```

```python
df['station_num'] = df.desc.str.split('Station', expand=True)[1].str.split(';', expand=True)[0]
df.head(5)
```

```python
df.station_num.str.replace(':', '').value_counts()[:10]
```

```python
data = df.station_num.str.replace(':', '').value_counts()[:10]
plt.figure(figsize=(12, 8))
x = list(data.index)
y = list(data.values)
x.reverse()
y.reverse()

plt.title("Most Called Stations")
plt.ylabel("Station")
plt.xlabel("Number of calls")

plt.barh(x, y)
```

```python
df.title.nunique()
```

```python
df['Reason_category'] = df.title.str.split(':', expand=True)[0]
df['Reason'] = df.title.str.split(':', expand=True)[1].str.replace(' -', '')
```

```python
df.head(5)
```

```python
df.Reason_category.value_counts()
```

```python
df.Reason.value_counts()
```

```python
data = df.Reason.value_counts()[:10]
plt.figure(figsize=(12, 8))
x = list(data.index)
y = list(data.values)
x.reverse()
y.reverse()

plt.title("Most emergency reasons of calls")
plt.ylabel("Reason")
plt.xlabel("Number of calls")

plt.barh(x, y)
```

```python
sns.countplot(x=df.Reason_category, data=df)
```

```python
df['timestamp'] = pd.to_datetime(df.timeStamp)
```

```python
df.timeStamp
```

```python
df['Hour'] = df.timestamp.dt.hour
df['Month'] = df.timestamp.dt.month
df['DayOfWeek'] = df.timestamp.dt.weekday
```

```python
df.head(5)
```

```python
df.Month.value_counts()
```

```python
sns.countplot(x=df.Month, data=df)
```

```python
df.Hour.value_counts()
```

```python
sns.countplot(x=df.Hour, data=df)
```

```python
dmap = {0:'Mon', 1:'Tue', 2:'Wed', 3:'Thu', 4:'Fri', 5:'Sat', 6:'Sun'}
df['DayOfWeek'] = df.DayOfWeek.map(dmap)
df.DayOfWeek.value_counts()
```

```python
sns.countplot(x=df.DayOfWeek, data=df, hue='Reason_category')
```

```python
byMonth = df.groupby('Month').count()
byMonth
```

```python
byMonth.twp.plot(figsize=(10, 8))
```

```python
plt.figure(figsize=(12, 8))
sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```

```python
df['Date'] = df.timestamp.dt.date
df.Date
```

```python
byDate = df.groupby("Date").count()
byDate.head()
```

```python
byDate.twp.plot(figsize=(10, 8))
```

```python
plt.figure(figsize=(15, 15))

for i, reason in enumerate(df.Reason_category.unique(), 1):
    plt.subplot(2, 2, i)
    df[df['Reason_category']==reason].groupby('Date').count()['twp'].plot()
    plt.title(reason)
    plt.tight_layout()
```

```python
dayHour = df.groupby(by=['DayOfWeek','Hour']).count()['Reason_category'].unstack()
dayHour.head()
```

```python
sns.heatmap(dayHour)
```

```python
sns.clustermap(dayHour)
```

```python
dayMonth = df.groupby(by=['DayOfWeek','Month']).count()['Reason_category'].unstack()
```

```python
sns.heatmap(dayMonth)
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(df.lng, df.lat, c=df.Month, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Month')
```
